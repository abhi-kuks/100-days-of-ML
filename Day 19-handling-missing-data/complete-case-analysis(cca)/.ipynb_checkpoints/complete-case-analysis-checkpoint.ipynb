```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_science_job.csv')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().mean()*100
```

```python
# So now we can do CCA{Complete Case Analysis} only on columns
# which have missing values lesser than 5%

cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() >0]
cols
```

```python
df[cols].sample(5)
```

```python
len(df)
```

```python
len(df[cols].dropna())
```

```python
# How much data will be left after removing all rows with nan values
# for columns having less than 5% {cols}

len(df[cols].dropna()) / len(df) *100
```

***So 89% of data will remain when we'll remove the NaN/null values***

```python
new_df = df[cols].dropna()
df.shape , new_df.shape
```

```python
new_df.hist(bins = 50  ,density = True , figsize = (12,12) )
plt.show()
```

```python
df[cols]._get_numeric_data().columns
```

```python
%matplotlib inline
```

```python
for column in list(df[cols]._get_numeric_data().columns):
    
    leg = {'Before CCA':'red' , 'After CCA':'green'}
    figs ,ax = plt.subplots(1, figsize = (12,12))
    df[column].hist(bins = 50 , ax= ax , color = 'red').legend(leg)
    new_df[column].hist(bins = 50 , ax= ax , color = 'green').legend(leg)
    plt.title(f'{column} Histogram before and After CCA')
    plt.show()
    plt.clf()
```

***We can clearly see that the distributions of numerical columns remains same throughout even after dropping the rows.***

```python
for column in list(df[cols]._get_numeric_data().columns):
    
    leg = {'Before CCA':'red' , 'After CCA':'green'}
    figs ,ax = plt.subplots(1, figsize = (12,12))
    df[column].plot.density(ax= ax , color = 'red').legend(leg)
    new_df[column].plot.density(ax =ax ,color = 'green').legend(leg)
    plt.title(f'{column} Histogram before and After CCA')
    plt.show()
    plt.clf()
```

***The smooth form of Histogram{PDF} is also coming similar, now this proves that the data was MCAR{Missing Completely at random}*** 

## Now let's talk about categorical data

```python
df['enrolled_university'].value_counts()
```

```python
df['education_level'].value_counts()
```

***So the ratio of categories of categorical date should be same after the CCA, which is coming out to be similar.***

```python
numerical =  list(df[cols]._get_numeric_data().columns)
categorical = list(set(df[cols].columns) - set(numerical))
print(categorical)
```

```python
uni = pd.concat([
        # Percenteage of observations per category original data
        df[categorical[0]].value_counts()  / len(df),

        #Percenteage of observations per category cca data
        new_df[categorical[0]].value_counts() / len(new_df)
    ] , axis =1 )

uni.columns = ['original' ,'cca']
```

```python
uni
```

```python
edu = pd.concat([
        # Percenteage of observations per category original data
        df[categorical[1]].value_counts()  / len(df),

        #Percenteage of observations per category cca data
        new_df[categorical[1]].value_counts() / len(new_df)
    ] , axis =1 )

edu.columns = ['original' ,'cca']
edu
```
